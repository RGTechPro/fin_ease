```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import joblib

# Load your dataset (replace this with your dataset)
data = pd.read_csv('your_file_transformed.csv')  # Replace 'your_dataset.csv' with the actual dataset file

# Selecting features and target variables
X = data.drop(['average_transaction_amount', 'frequent_transaction_modes'], axis=1)  # Features
y_average = data['average_transaction_amount']  # Target variable for average_transaction_amount
y_modes = data['frequent_transaction_modes']  # Target variable for frequent_transaction_modes

# Splitting the dataset into training and testing sets
X_train, X_test, y_average_train, y_average_test, y_modes_train, y_modes_test = train_test_split(
    X, y_average, y_modes, test_size=0.2, random_state=42
)

# Create and train the individual models
logreg_model = LogisticRegression(max_iter=1000, random_state=42)
logreg_model.fit(X_train, y_average_train)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_modes_train)

gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_modes_train)

# Create a Voting Classifier
voting_classifier = VotingClassifier(estimators=[
    ('logreg', logreg_model),
    ('random_forest', rf_model),
    ('gradient_boosting', gb_model)
], voting='soft')  # 'soft' uses predicted class probabilities for voting

# Train the Voting Classifier
voting_classifier.fit(X_train, y_modes_train)

joblib.dump(voting_classifier, 'voting_classifier.pkl')

# Make predictions using the Voting Classifier
y_pred = voting_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_modes_test, y_pred)
print(f"Accuracy of the Voting Classifier: {accuracy:.2f}")
```
